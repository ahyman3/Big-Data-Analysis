# Finding which model works best in Arkansas

```python
#Modeling all arkansas cities with ARIMA model
#the format will be a tuple in the shape of:
#(city, order, mse)
#Tuple with the best MSE for each of the cities will be 
#stored in arkansas_arima_mse list
arkansas_arima_mse = []
#d values
i = [1,2]
#p values for AR
arp = [1,2,3,4]
#q values for MA
maq = [1,2,3,4]
#For each of the cities
for city in all_arkansas.columns:
    #Index the training data
    train = all_arkansas[city]["01-01-2007":"12-31-2017"]
    #Index the test data
    test = all_arkansas[city]["2018":]
    #For each permutation
    for p in arp:
        for q in maq:
            for d in i:
                #Create the order to model
                order = (p, d, q)
                try:
                    #Mode the training data with the particular order
                    arima_temp = ARIMA(train, order=order).fit()
                    #Forecast the data nine periods
                    f, err, ci = arima_temp.forecast(9)
                    #Calculat the MSE
                    mse = mean_squared_error(y_pred=f, y_true=test)
                    #Store the information in the list
                    arkansas_arima_mse.append((city, order, mse))
                except:
                    pass
```

```python
sorted(arkansas_arima_mse, key = lambda t: t[2])
lr_mse = [data for data in arkansas_arima_mse if data[0] == "Little Rock"]
lr_mse = sorted(lr_mse, key=lambda d: d[2])
hs_mse = [data for data in arkansas_arima_mse if data[0] == "Hot Springs"]
hs_mse = sorted(hs_mse, key=lambda d: d[2])
fv_mse = [data for data in arkansas_arima_mse if data[0] == "Fayetteville"]
fv_mse = sorted(fv_mse, key=lambda d: d[2])
sr_mse = [data for data in arkansas_arima_mse if data[0] == "Searcy"]
sr_mse = sorted(sr_mse, key=lambda d: d[2])

print(lr_mse[:10])
print(hs_mse[:10])
print(fv_mse[:10])
print(sr_mse[:10])
```

```python
plt.rcParams["figure.figsize"] = (12,16)
plt.subplot(411)
#train
fville_train = fville["01-01-2007":"12-31-2017"]
#test
fville_test = fville["01-01-2018":]

#Playing through the froecasting of the arkansas cities
arima_model = ARIMA(fville_train, order = (3,2,1)).fit(trend = "nc")

#Forecast
f, err, ci = arima_model.forecast(9)

#Forecast into a data frame
f = pd.DataFrame(index = fville_test.index, data= {"price": f})
#Confidence interval into a data frame
ci = pd.DataFrame(index = fville_test.index, data= {"ub": ci[:,0], "lb":ci[:,1]})


#plotting the 
plt.plot(fville_test)
plt.plot(f)
plt.plot(ci)
plt.title("Fayetteville Forecast ARIMA(3,2,1)")
plt.legend(["Actual", "Forecast", "Lower Bound", "Upper Bound"])

rmse_fville = np.sqrt(mean_squared_error(y_pred=f, y_true=fville_test))

plt.subplot(412)
#train
hspring_train = hspring["01-01-2007":"12-31-2017"]
#test
hspring_test = hspring["01-01-2018":]

#Playing through the froecasting of the arkansas cities
arima_model = ARIMA(hspring_train.tolist(), order = (1,2,2)).fit(trend = "nc")

#Forecast
f, err, ci = arima_model.forecast(9)

#Forecast into a data frame
f = pd.DataFrame(index = hspring_test.index, data= {"price": f})
#Confidence interval into a data frame
ci = pd.DataFrame(index = hspring_test.index, data= {"ub": ci[:,0], "lb":ci[:,1]})


#plotting the 
plt.plot(hspring_test)
plt.plot(f)
plt.plot(ci)
plt.title("Hot Springs Forecast ARIMA(1,2,2)")
plt.legend(["Actual", "Forecast", "Lower Bound", "Upper Bound"])

rmse_hs = np.sqrt(mean_squared_error(y_pred=f, y_true=hspring_test))

plt.subplot(413)
#test
searcy_train = searcy["2007":"2017"] 
searcy_test = searcy["01-01-2018":]

#Playing through the froecasting of the arkansas cities
arima_model = ARIMA(searcy_train.tolist(), order = (2,1,4)).fit(trend = "nc")

#Forecast
f, err, ci = arima_model.forecast(9)

#Forecast into a data frame
f = pd.DataFrame(index = searcy_test.index, data= {"price": f})
#Confidence interval into a data frame
ci = pd.DataFrame(index = searcy_test.index, data= {"ub": ci[:,0], "lb":ci[:,1]})


#plotting the 
plt.plot(searcy_test)
plt.plot(f)
plt.plot(ci)
plt.title("Searcy Forecast ARIMA(2,1,4)")
plt.legend(["Actual", "Forecast", "Lower Bound", "Upper Bound"])

rmse_searcy = np.sqrt(mean_squared_error(y_pred=f, y_true=searcy_test))

plt.subplot(414)
#train
lr_train = littlerock["01-01-2007":"12-31-2017"]
#test
lr_test = littlerock["01-01-2018":]

#Playing through the froecasting of the arkansas cities
arima_model = ARIMA(lr_train.tolist(), order = (1,1,1)).fit(trend = "nc")

#Forecast
f, err, ci = arima_model.forecast(9)

#Forecast into a data frame
f = pd.DataFrame(index = lr_test.index, data= {"price": f})
#Confidence interval into a data frame
ci = pd.DataFrame(index = lr_test.index, data= {"ub": ci[:,0], "lb":ci[:,1]})


#plotting the Little Rock forecast
plt.plot(lr_test)
plt.plot(f)
plt.plot(ci)
plt.title("Little Rock Forecast ARIMA(1,1,1)")
plt.legend(["Actual", "Forecast", "Lower Bound", "Upper Bound"])

plt.show()

rmse_lr = np.sqrt(mean_squared_error(y_pred=f, y_true=lr_test))



print("Fayetteville RMSE: {:.2f}".format(rmse_fville))
print("Hot Springs RMSE: {:.2f}".format(rmse_hs))
print("Searcy RMSE: {:.2f}".format(rmse_searcy))
print("Little Rock RMSE: {:.2f}".format(rmse_lr))
```

```python
filtered_homes = cheap_homes[filtered_zips]
filtered_agi = cheap_agi[filtered_zips]
```

```python
#Home values for training
homes_v = filtered_homes["01-01-2018":]
homes_tr = filtered_homes["01-01-2011":"12-01-2016"]
homes_test = filtered_homes["01-01-2017":"12-01-2017"]

#AGI data for training 
agi_v = filtered_agi["01-01-2018":]
agi_tr = filtered_agi["01-01-2011":"12-01-2016"]
agi_test = filtered_agi["01-01-2017":"12-01-2017"]
```

```python
#Getting all zip codes to test
all_zips = homes_v.columns
```

```python
#MA to sort through
ma = [1,2,3,4]
#Differencing to sort through
i = [1,2]

#List to hold a tuple of the city name and order with best bic
best_bic_list = []
#List to hold a tuple of the city name and order with best rmse
best_rmse_list = []

#For each of the remaining zip codes
for zc in all_zips:
    #Iniitalize no best bic for the zip code
    best_bic = None
    #Iniitalize no best bic order for the zip code
    best_bic_order = None
    #Iniitalize no best rmse for the zip code
    best_rmse = None
    #Iniitalize no best rmse order for the zip code
    best_rmse_order = None
    #Testing both first order and second order differencing
    for d in i:
        #Testing 1-4 order MA
        for q in ma:
            try:
                #model uses the 2011-2017 data for zip code
                #model uses the 2011-2017 data for AGI
                #Trying each of the order provided in the loop
                #Not applying any extra model on the season
                #Fitting the non-constant trend
                model = SARIMAX(homes_tr[[zc]], agi_tr[[zc]], order = (1, d, q), seasonal_order=(0,0,0,12)).fit(trend = "nc")
                #Getting the forecast for 12 periods and interpolated AGI
                f = model.get_forecast(12, exog = agi_test[[zc]])
                #If the bic is less than the best BIC
                if best_bic is None or best_bic > model.bic:
                    #Save the order/bic
                    best_bic = model.bic
                    best_bic_order = (1,d,q)
                #Calculating rmse
                rmse = np.sqrt(mean_squared_error(y_pred=f.predicted_mean, y_true=homes_test[zc]))
                #If best RMSE for the zip code
                if best_rmse is None or best_rmse > rmse:
                    #Save order/rmse
                    best_rmse = rmse
                    best_rmse_order = (1,d,q)
            #For when it is not stationary
            except:
                pass
    #Put the best bic and rmse order in the list 
    best_bic_list.append((zc, best_bic_order))
    best_rmse_list.append((zc, best_rmse_order))
    
                
```

```python
#Dictionary to organize the data
bic_dict = {}
#Getting the order and placing in list to count frequency later
bic_list = [bic[1] for bic in best_bic_list]
#Repeat for RMSE
rmse_dict = {}
rmse_list = [rmse[1] for rmse in best_rmse_list]
#Counting order that resulted in best BIC
for bic in bic_list:
    bic_dict[bic] = bic_dict.get(bic, 0) + 1
#Counting order that resulted in best RMSE
for rmse in rmse_list:
    rmse_dict[rmse] = rmse_dict.get(rmse, 0) + 1
```

```python
#putting the dictioary items in a list so I can sort
bic_items = [bic for bic in bic_dict.items()]
#Sorting to see the orders that had the best bic first
bic_items = sorted(bic_items, key = lambda d: d[1], reverse=True)
#Putting rmse orders in a list to sort
rmse_items = [rmse for rmse in rmse_dict.items()]
#sorting the rmse by the order that had resulted in the best rmse the most times
rmse_items = sorted(rmse_items, key = lambda d: d[1], reverse=True)
print("Best BIC:\n", bic_items)
print("Best RMSE:\n", rmse_items)
```

```python
#The home prices 2011-2017
homes_tr = filtered_homes["01-01-2011":"12-01-2017"]

#AGI 2011-2017
agi_tr = filtered_agi["01-01-2011":"12-01-2017"]
#Intrpolated AGI 2011-2017
agi_v = filtered_agi["01-01-2018":"12-01-2018"]
```

```python
forecast_list = []
for item in best_bic_list:
    zip_dict = {}
    #unpacking the order
    col, order = item
    try:
        #modeling
        model = SARIMAX(homes_tr[[col]], agi_tr[[col]], order = order, seasonal_order=(0,0,0,12)).fit(trend = "nc")
        #getting forecast
        f = model.get_forecast(12, exog = agi_v[[col]])
        #getting all lowerbound values
        lower = f.conf_int().values[:,0]
        #Getting the final lower bound prediction
        last_lower = lower[-1]
        #getting all the upper confidence interval
        upper = f.conf_int().values[:,1]
        #Getting all the predicted values
        predicted = f.predicted_mean
        #getting the last predicted value
        last_predicted = f.predicted_mean[-1]
        #getting the last known value
        last_known = homes_tr[col][-1]
        #calcuating the return on the real estate
        forecasted_pc = ((last_predicted - last_known) / last_known) * 100
        #Getting lower bound forecast
        lowerbound_pc = ((last_lower - last_known) / last_lower) * 100
        #Standardized conf interval
        confint = (upper[-1] - lower[-1]) / predicted[-1] * 100
        #Dictionary for city
        zip_dict["zip"] = col
        zip_dict["forecasted_pc"] = forecasted_pc
        zip_dict["lower_pc"] = lowerbound_pc
        zip_dict["confIntNorm"] = confint
        forecast_list.append(zip_dict)
    except:
        pass
```

```python
#Putting forecasts in a data frame
forecasts = pd.DataFrame(data = forecast_list)
```
